```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Payment Tier 1 (Highest), 2 (Mid Level), 3 (Lowest)
# Experience 1 - 7 

```python
df = pd.read_csv('Employee.csv')
df.head()
```

# Grafik Batang
# Pada grafik batang di bawah, kita dapat melihat perbandingan lokasi tempat dimana karyawan bekerja.
## Grafik ini membagi kota dalam 3 kategori : Bangalore, Pune, dan New Delhi.
# Bangalore : Sebanyak lebih dari 2000 karyawan bekerja disini. Bangalore termasuk lokasi dengan jumlah karyawan terbanyak dalam grafik. (Grafik Hijau)
# Pune : Sekitar 1250 karyawan bekerja di Pune. (Grafik Merah Muda)
# New Delhi : Sekitar 1150 karyawan bekerja di New Delhi. (Grafik Biru)

```python
dfCity = df['City'].value_counts()

plt.figure(figsize=(15,10))
plt.bar(dfCity.index,dfCity,color=['palegreen','pink','skyblue'])
plt.xlabel('Kota', fontweight='bold')
plt.ylabel('Jumlah', fontweight='bold')
plt.title('Perbandingan Jumlah Kota Karyawan', fontsize=20, fontweight='bold')
plt.show() 
```

# Histogram
# Pada histogram di bawah, kita dapat melihat perbandingan usia karyawan dalam perusahaan tersebut dengan membagi kedalam 7 interval.
# Usia 22.5 - 25 : lebih dari 400 karyawan
# Usia 25 - 27.5 : lebih dari 1600 karyawan
# Usia 27.5 - 30 : lebih dari 1000 karyawan
# Usia 30 - 32.5 : lebih dari 200 karyawan
# Usia 32.5 - 35 : sekitar 400 karyawan
# Usia 35 - 37.5 : lebih dari 400 karyawan
# Usia 37.5 - 40 : lebih dari 200 karyawan

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'],color='pink', bins=7)
plt.xlabel('Umur', fontweight='bold')
plt.ylabel('Jumlah', fontweight='bold')
plt.title('Usia Karyawan', fontsize=20, fontweight='bold')
plt.show()
```

# Pie Chart
# Pada diagram ini, kita dapat melihat tingkat pendidikan karyawan dalam perusahaan. Dibagi dalam 3 kategori yaitu Bachelors, Masters, dan PHD
# Bachelors : sebanyak 77.4% dari karyawan
# Masters : sebanyak 18.8% dari karyawan
# PhD : Sebanyak 3.8% dari karyawan

```python
plt.figure(figsize=(5,5))
dfEdu = df['Education'].value_counts()
plt.pie(dfEdu, labels=dfEdu.index,autopct="%1.1f%%", startangle=90, shadow=True)
plt.title('Tingkat Pendidikan Karyawan', fontweight='bold')
plt.show()
```

# Heatmap
# Pada heatmap hanya data yang bersifat numerik yang ditampilkan korelasinya
# Korelasi Positif : Warna merah - kuning dengan nilai mendekati 1 (nilai satu variabel meningkat, nilai variabel lain meningkat)
# Korelasi Negatif : Warna pink - ungu dengan nilai mendekati -1 (nilai satu variabel meningkat, nilai variabel lain menurun)
# Korelasi Nol : warna terang (mendekati putih) dan nilai mendekati 0 menunjukan tidak ada relasi antar kedua variabel
# Pada data berikut, dapat dilihat bahwa 
# Joining Year dengan Payment Tier berkorelasi negatif : jika karyawan bergabung awal cenderung pendapatannya banyak
# Payment Tier dengan Age berkolerasi positif rendah : jika semakin tua umur karyawan maka pendapatannya sedikit
# Age dengan ExperienceInCurrentDomain berkolerasi negatif : jika semakin tua umur karyawan maka sedikit pengalaman dalam bidangnya
# LeaveOrNot dengan Age berkolerasi negatif : jika semakin tua usia karyawan maka keinginan untuk meninggalkan perusahaan rendah
# ExperienceInCurrentDomain dengan Payment Tier berkolerasi positif rendah : jika semakin tinggi pengalaman dalam bidangnya maka pendapatannya banyak

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(), 
            annot=True)
plt.title('Heatmap', fontsize=20)
plt.show()
```
